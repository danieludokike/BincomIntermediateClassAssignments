```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

### Reading load data to a dataframe

```python
loan_df = pd.read_csv("data/loan_data.csv")
```

```python
loan_df
```


## Handling missing values

```python
loan_df.fillna(method='ffill', inplace=True)
```

```python
loan_df
```

## Replacing 3+ with 3 in the Dependents column

```python
loan_df['Dependents'] = loan_df['Dependents'].replace({'3+': 3})
```

## Encoding for Preprocessing techniques for categorical to numerical vals converstions

```python
# encoder = OneHotEncoder()
# encoded_vals = encoder.fit_transform(loan_df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']])
loan_df = pd.get_dummies(loan_df, columns=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area'])
```

## Spliting into features and target val for evalutations

```python
x_vals = loan_df.drop(['Loan_ID', 'Loan_Status'], axis=1)
y_vals = loan_df['Loan_Status']
```

## Training and testing sets and train with RandomForestClassifier

```python
x_vals_train, x_vals_test, y_vals_train, y_vals_test = train_test_split(x_vals, y_vals, test_size=0.2, random_state=42)

# model = LogisticRegression()
# model.fit(x_vals_train, y_vals_train)

# Using HistGradientBoostingClassifier since LogisticRegression can't handle NaN native values
from sklearn.ensemble import HistGradientBoostingClassifier
model = HistGradientBoostingClassifier()
model.fit(x_vals_train, y_vals_train)
```

## Making predictions

```python
predictions_on_y = model.predict(x_vals_test)
```

## Model evaluation and classification report

```python
accuracy = accuracy_score(y_vals_test, predictions_on_y)
print(f'Accuracy: {accuracy}')
```

## Making prediction with a new data to test our  model

```python
test_data = pd.DataFrame({
    'Gender': ['Male'],
    'Married': ['Yes'],
#     'Dependents': [1],
    'Education': ['Graduate'],
    'Self_Employed': ['No'],
#     'ApplicantIncome': [5000],
#     'CoapplicantIncome': [2000],
#     'LoanAmount': [150],
#     'Loan_Amount_Term': [360],
#     'Credit_History': [1],
    'Property_Area': ['Urban']
})
test_data = pd.get_dummies(test_data)
test_data_prediction = model.predict(test_data)
print(f'Loan Status Prediction: {test_data_prediction}')
```
